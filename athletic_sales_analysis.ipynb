```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
athletic_sales_2021_path = Path('Resources/athletic_sales_2021.csv')

athletic_sales_2020_df = pd.read_csv(athletic_sales_2020_path, index_col="retailer")
athletic_sales_2021_df = pd.read_csv(athletic_sales_2021_path, index_col="retailer")
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", join="inner")
joined_data_rows
```

```python
# Check if any values are null.
clean_athletic_df = athletic_sales_2021_df.dropna(how="any")
clean_athletic_df.count()
```

```python
# Check the data type of each column
clean_athletic_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
clean_athletic_df['invoice_date']= pd.to_datetime(clean_athletic_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
clean_athletic_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
region_counts = clean_athletic_df["region"].value_counts()
region_counts.head()

grouped_clean_athletic_df = clean_athletic_df.groupby(['region','state','city'])
# Rename the sum to "Total_Products_Sold".


# Show the top 5 results.
grouped_clean_athletic_df.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
product_region_sum = pd.pivot_table(clean_athletic_df,
index=['region', 'state', 'city'],
values='units_sold',
aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
product_region_totals = product_region_sum.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
product_region_sum.head(5)
```
